### Task 2: Add Month Column

```python
df['Month'] = df['Order Date'].str[0:2]
df
```

#### coverting column to the right type

```python
df.loc[:,'Quantity Ordered'] = df.loc[:,'Quantity Ordered'].astype(int)
df.loc[:,'Price Each'] = df.loc[:,'Price Each'].astype(float)
df.head()
```

### Task 3: Add Sales Column

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

### Task 4: Add a City Column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
df
```

#### Question 4: What products are most often sold together?

```python
df2 = df[df['Order ID'].duplicated(keep = False)]
df2['Grouped'] =  df2.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df2[['Order ID','Grouped']].drop_duplicates()
df2.head(10)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

#### Question 6: What product sold the most? Why do you think it sold the most?

```python
product_group = df.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.title('The Most Sold Product')
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.xticks(products, rotation = 'vertical', size = 8)

plt.show()
```

```python
prices =  df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color ='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('Prices ($)', color = 'b')
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)
ax1.set_title('Product and Price Correlation')

plt.show()
```
